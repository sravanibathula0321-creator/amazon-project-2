````python
# Visualizing the missing values
# Set consistent plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

#  Function to plot missing values as a bar chart
def plot_missing_bar(df, title):
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values()

    if not missing.empty:
        plt.figure(figsize=(10, 5))
        missing.plot(kind='barh', color='crimson')
        plt.title(f" Missing Values Count - {title}")
        plt.xlabel("Number of Missing Values")
        plt.ylabel("Column")
        plt.tight_layout()
        plt.show()
    else:
        print(f" No missing values in {title}.")

# Function to plot missing values as a heatmap
def plot_missing_heatmap(df, title):
    if df.isnull().values.any():
        plt.figure(figsize=(12, 6))
        sns.heatmap(df.isnull(), cbar=False, cmap="magma", yticklabels=False)
        plt.title(f" Missing Data Heatmap - {title}")
        plt.xlabel("Columns")
        plt.ylabel("Rows with Missing Values")
        plt.tight_layout()
        plt.show()
    else:
        print(f" No missing values in {title}.")

# Visualize Missing Values in merged dataset (df)
plot_missing_bar(df, "Merged Dataset")
plot_missing_heatmap(df, "Merged Dataset")
````

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Step 1: Preview the data
print("Initial shape:", df.shape)
print(df.head())

# Remove duplicate rows from the original dataframes before merging if needed,
# or handle them in the merged dataframe as done below.
# For the merged df, decide which columns define a duplicate row.
# Assuming a duplicate is a row with the same title, person, and role:
# Using 'id' from titles_df and 'person_id', 'name', 'role', 'character' from credits_df
# When merged, 'id' from titles_df becomes the shared column.
# A potentially better subset for duplicates in the merged df might involve the content id and person id.
subset_cols_for_duplicates = ['id', 'person_id', 'role', 'character'] # Use relevant columns that uniquely identify a person's credit for a title. 'name' is redundant with 'person_id' here.
initial_rows = df.shape[0]
df = df.drop_duplicates(subset=subset_cols_for_duplicates)
print(f"\nRemoved {initial_rows - df.shape[0]} duplicate rows.")
print("\nShape after removing duplicates:", df.shape)


# Step 3: Strip whitespace from column names
df.columns = df.columns.str.strip()

# Step 4: Check and handle missing values
missing = df.isnull().sum()
print("\nMissing values before handling:\n", missing)

# Drop columns with too many nulls if any (more than 50%)
threshold = len(df) * 0.5
# Check for columns to drop before dropping to avoid errors if none meet criteria
cols_to_drop = missing[missing > threshold].index.tolist()
if cols_to_drop:
  df = df.drop(columns=cols_to_drop)
  print(f"Dropped columns with >50% missing values: {cols_to_drop}")

# Recalculate missing after dropping columns
missing = df.isnull().sum()

# Fill or drop remaining NaNs
# Ensure essential columns exist before dropping based on them
if 'title' in df.columns:
    df = df.dropna(subset=['title'])  # Title is essential
else:
    print("Warning: 'title' column not found. Skipping dropna based on 'title'.")

# Fill 'genres' column if it exists
if 'genres' in df.columns:
    df['genres'] = df['genres'].fillna('Unknown')

# Fill 'release_year' column if it exists and has missing values
if 'release_year' in df.columns and df['release_year'].isnull().sum() > 0:
    # Check if there's a mode before trying to access it
    if not df['release_year'].mode().empty:
        df['release_year'] = df['release_year'].fillna(int(df['release_year'].mode()[0]))  # Most frequent year, cast to int
    else:
         # If no mode (e.g., all NaN), fill with a default or handle differently
         print("Warning: 'release_year' column has no mode. Cannot fill with mode.")
         # Optionally fill with another value like a placeholder or overall median/mean
         # df['release_year'] = df['release_year'].fillna(-1) # Example placeholder

# The column name in titles_df is 'runtime'. Let's ensure we use that name.
# Check if 'runtime' exists before trying to fill NaNs.
if 'runtime' in df.columns:
    # Only fill if column exists and has missing values
    if df['runtime'].isnull().sum() > 0:
        df['runtime'] = df['runtime'].fillna('Unknown')
else:
    print("Warning: 'runtime' column not found in the DataFrame. Skipping filling NaNs for 'runtime'.")


# Handle missing 'character' and 'age_certification' based on previous plan
if 'character' in df.columns:
    df['character'] = df['character'].fillna('Unknown')
if 'age_certification' in df.columns:
    df['age_certification'] = df['age_certification'].fillna('Not Rated')

# Fill numerical missing values if needed (using median as in the description)
# Identify numerical columns remaining after dropping high-missing columns
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
for col in ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']:
    # Check if column exists and has missing values before trying to fill
    if col in numerical_cols and df[col].isnull().sum() > 0:
        median_val = df[col].median()
        # Check if median is not NaN (can happen if all values in column are NaN)
        if pd.notnull(median_val):
            df[col] = df[col].fillna(median_val)
            print(f"Filled missing values in '{col}' with median: {median_val}")
        else:
             print(f"Warning: Median for '{col}' is NaN. Cannot fill missing values.")


# Check missing values after handling
print("\nMissing values after handling:\n", df.isnull().sum()[df.isnull().sum() > 0])


# Step 5: Convert data types
# Example: release_year to int (already handled during fillna, but ensure type is int)
if 'release_year' in df.columns:
    try:
        # Use pd.to_numeric with errors='coerce' to turn non-convertible values into NaN
        # Then fill NaNs if needed before converting to int
        df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
        # Fill any NaNs created by coercion before casting to int
        if df['release_year'].isnull().sum() > 0:
             # Decide how to fill: median, mode, or a placeholder
             print("Warning: Non-integer values found in 'release_year', converted to NaN.")
             # For simplicity here, filling with mode if available, otherwise a placeholder like -1
             if not df['release_year'].mode().empty:
                 fill_value = int(df['release_year'].mode()[0])
                 df['release_year'] = df['release_year'].fillna(fill_value)
                 print(f"Filled new NaNs in 'release_year' with mode: {fill_value}")
             else:
                  df['release_year'] = df['release_year'].fillna(-1) # Use a clear placeholder
                  print("Warning: No mode found for 'release_year'. Filled new NaNs with -1.")

        df['release_year'] = df['release_year'].astype(int) # Now cast to int
    except Exception as e:
        print(f"Could not convert 'release_year' to int after handling NaNs: {e}")


# Step 6: Clean string columns
# 'duration' does not seem to be a column in the merged dataframe based on the original datasets.
# Let's use the actual column names found after merging ('type', 'genres', 'title', 'runtime', 'character', etc.)
str_cols_to_clean = ['title', 'type', 'character', 'age_certification']
if 'runtime' in df.columns:
    str_cols_to_clean.append('runtime') # Add runtime if it exists
# Ensure 'name' column from credits is also cleaned if it's a string
if 'name' in df.columns:
     str_cols_to_clean.append('name')


# Use a more robust way to identify string columns if needed
# object_cols = df.select_dtypes(include='object').columns.tolist()
# str_cols_to_clean = [col for col in object_cols if col not in ['genres', 'production_countries']] # Exclude list-like columns

for col in str_cols_to_clean:
    # ADD THIS CHECK: Ensure the column exists in the dataframe before processing
    if col in df.columns:
        # Ensure the column is of object type before applying string methods
        # This check prevents errors on numerical columns accidentally added to the list
        if pd.api.types.is_object_dtype(df[col]):
           df[col] = df[col].astype(str).str.strip().str.title()
        else:
            print(f"Warning: Column '{col}' is not of object dtype. Skipping string cleaning.")
    else:
        print(f"Warning: Column '{col}' not found in DataFrame. Skipping string cleaning.")


# Step 7: Normalize 'genres' (optional: multiple genres split into list)
# The goal here seems to be splitting comma-separated genres into a list.
# Ensure the column exists and is treated as a string before splitting.
if 'genres' in df.columns:
     # Convert to string first to handle potential non-string entries (like 'Unknown')
     df['genres'] = df['genres'].astype(str)
     # Use a lambda function with error handling in case of unexpected formats
     # Ensure 'Unknown' string is handled correctly if present
     df['genres'] = df['genres'].apply(lambda x: [g.strip().title() for g in x.replace(" ", "").split(',') if g and g != 'Unknown'] if pd.notnull(x) and x != 'Unknown' else ['Unknown'])
     # The previous apply might create empty lists for 'Unknown' or NaNs.
     # Let's ensure ['Unknown'] is used for empty or original 'Unknown' cases.
     df['genres'] = df['genres'].apply(lambda x: ['Unknown'] if not x or x == [''] else x) # Use [''] to catch potential empty strings from split

else:
    print("Warning: 'genres' column not found. Skipping genre normalization.")


# Step 8: Check final info
print("\nFinal dataset info:")
print(df.info())

# Display final missing values count to confirm
print("\nFinal Missing Values Count:")
print(df.isnull().sum()[df.isnull().sum() > 0])
````
